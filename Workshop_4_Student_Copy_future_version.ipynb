# For Loops Continued:

Computers are very good at clearly-outlined, repetitive tasks. "For Loops" are one case of these - you tell Python go to through an object, and for every element, to execute a block of code, which that object may or may not turn up in.

Their basic structure:

```
for i in x:
    [do something]
  
```

They're called "for" loops, because if you changed the above to English, we'd say "for every element `i` in object `x`, do something." `i` changes each time it loops through, equal to the first, then the second, then third, etc, element of the `x` object. 

Here's a silly example (the functions are all made up!) of something most of us probably do:
```
week = ["Monday", "Tuesday", "Wednesday", "Thursday", "Friday", "Saturday", "Sunday"]

for morning in week:
    eat_breakfast("Toast")
    brush_teeth()
    get_dressed()
```



Here's some examples with actual Python code


### Simple For Loop Examples

````python
for character in "PYTHON":
  print(character)
````

You can loop through strings like "PYTHON" above since they are an iterable data type. For now, the most important iterable data type is the list. Dictionaries are also iterable, but it's more complex. 

So what the above does is first, it sets `character` equal to the first element of `"PYTHON"`. Then it runs the code block, `print(character)`, and prints "P". 

That iteration terminates, since there's nothing else to do. Then python goes back to the start, and sets `character` equal to the second letter in `"PYTHON"`, and runs the code block.

Python repeats this for every element in `"PYTHON"`. Once it's out of elements to iterate through, the loop terminates. 

This is called a `for` loop, because it does an operation `for` every element of the object it's iterating through.

````python
for character in 12345:
  print(character)
````

This doesn't work! Integers aren't iterable, since they have meaning as one entity. We can make the word "Python" by adding all of its letters together in order, but 12345 is not equal to 1 + 2 + 3 + 4 + 5!

What this looks like done right:

````python
for integer in [1,2,3,4,5]:
  print(integer)
````

This program, in English, might be phrased as "for every integer from 1 to 5, say that integer."

The `integer` value is first set to the value of the first element of the list, 1. Then it executes the code block under those circumstances - `integer` equals 1, so it prints 1, then "loops" back to the beginning, moving on to the next value of the list.

Now `integer` is set to the next value of the list, 2, and executes the code block again. It prints 2, and loops back to the start. 

Like this, it prints 3, 4, and 5. At 5, there are no longer any more elements left in the list, so the program terminates.

This case is simple enough. We could just line up five `print()` statements. But for complex, repetitive operations, you'll often prefer to use a loop.

### For Loop: Indexing Through List 

One very useful case involves indexing through lists. As we know, every element in a list has an index number, starting at 0 for the first element and counting up by 1. 

Sometimes, you want to access every element of a list in order, do an operation, and return the value. 

We can do this with the following: 

````python
for index in range(len(gdp)):
  print(index)
````

````python
for index in range(len(gdp)):
  print(2*gdp[index])
````

Let's break this down line by line.

`for index in range(len(gdp)):`

We use two functions on the `gdp` variable. The first, `len()`, we are familiar with - it takes an object and returns how many elements it holds. Since `gdp` has 4 elements, it returns 4. 

The second function, `range()`, creates a range of integers the length of the input value. When you call `range(4)` it doesn't return `[0, 1, 2, 3]`, but `range(0,4)`. On its own, it's not that useful.

But when you iterate through it in a list, as below, it does what we want. 


````python
range(4)
````

````python
for index in range(len(gdp)):
  print(index)
````

Now let's take a look at the second part of the for loop.

```
for index in range(len(gdp)):
   print(2*gdp[index])
```

We know `print()` - it's telling Python to print whatever value we get. 

The input is using `index`. Since this is drawn from the `range(len(gdp))` value, `index` is set to 0, 1, 2, or 3, accesses that element from `gdp`, and multiplies it by 2.

So, the overall loop is saying, **"for every value in the gdp list, multiply it by 2, and print this value."**

### Modular Code

*   `range(len(gdp)` is more complicated than simply inputting  [0, 1, 2, 3]
*   Why not use the manual list?

By building the range we iterate over from the object (`gdp`) that we want to work on, if we change the number of elements in `gdp`, we can still use the same code. 

As you get familiar with python, you'll be able to read wrapped functions like `range(len(gdp))` more easily.

# More Complicated

When you create a function, usually you'll want have more than one argument (arguments are like input variables).

When you have multiple arguments, all of them must be given a value when you use the function in code. If you don't specify which is which, Python will assume they're in the same order as when the function was assigned.

````python
def my_function_2args(fname, lname):
  """
  Takes `fname` and `lname`, both strings, and prints a welcome message.
  """
  print("Hello " + fname + " " + lname + ". Welcome to class!")

my_function_2args("Harry", "Izatt")
````

An exception is when functions have default arguments

````python
def my_function_default_args(fname, lname="Izatt"):
  """
  Takes `fname` and `lname`, both strings, and prints a welcome message. 
  `lname` defaults to "Izatt"
  """
  print("Hello " + fname + " " + lname + ". Welcome to class!")
````

````python
print("These first two get the same output, due to the default `lname` variable.")
my_function_default_args("Harry")
my_function_default_args(fname = "Harry", lname = "Izatt")

print(" ")

print("If we forget an `lname`, it might not always be correct...")
my_function_default_args(lname = "Trudeau", fname = "Justin")
my_function_default_args(fname = "Justin")

print(" ")

print("Or if we put arguments out of order without specifying which they are!")
my_function_default_args("Trudeau", "Justin")
````

Compare the third and fifth usages of `my_function_default_args()`. Both of them put a last name first. But in the third we specify which argument is which, so it works fine. In the fifth we forgot to, so the arguments are out of the order we want. 

````python
my_function_default_args(fname = "Harry", lname = "Izatt")
````

# Functions: To Return or Not To Return

In a function, you can include a statement at the end formatted as:

`
  return ____
`

In this case, the function will be equal to whatever is in the `____` space. So, you can do a statement like:

`
complete_dataframe = data_cleaning_function(datasource)
`

where `data_cleaning_function()` takes the `datasource` variable, does something to it, and returns some value for the cleaned data. We assign this as `complete_dataframe`

Alternatively, if we just want to print a message, or don't need to create and store a variable, we might not want to include the return statement. 

This is a more advanced concept - you'd easily spend a week or more on learning how to make more elaborate functions using `return ____`, in a real course. Many functions you'll install from other people are formatted this way, so this is important to know.

````python
def double_return(num_list):
  """
  Takes a list of numbers as `num_list`, doubles them, and returns the new list.
  """
  # creates empty list
  doubled_return_list = []

  # iterates through list
  for i in range(len(num_list)):
    # accesses num_list elements in order, doubles them, and adds them to the returned list
    doubled_return_list.append(2*num_list[i])
  
  return doubled_return_list
````

````python
double_return(gdp)
````

````python
doubled_gdp = double_return(gdp)
print(doubled_gdp)
````

````python
double_return(gdp)*2
````

````python
def double_print(num_list):
  """
  Takes a list of numbers as `num_list`, doubles them, and prints the new list.
  """

  # creates empty list
  doubled_print_list = []
  
  # iterates through list
  for i in range(len(num_list)):
    # accesses num_list elements in order, doubles them, and adds them to printed list
    doubled_print_list.append(2*num_list[i])
  
  print(doubled_print_list)
````

````python
double_print(gdp)
````

````python
# this much works...

doubled_gdp = double_print(gdp)
````

````python
# but this will throw an error:

double_print(gdp)*2
````

How to think of `return` versus `print()`

[A good explanation is here.](https://www.codecademy.com/forum_questions/518ffbfeb3f05c44fe001395#:~:text=print%20just%20shows%20the%20human,the%20computer%20in%20further%20functions.)

In essence, `print()` displays value for a human to see, but that the computer cannot interact with. `return` gives values for Python to continue operating with, but might also be viewable for the human.

Do you need to modify and keep working with values? Use `return`. Do you just want to see them? Use `print()`
